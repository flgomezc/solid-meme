```python
from datetime import datetime
```

```python
window_max = datetime.strptime('2021-01-20 12:00:00', '%Y-%m-%d %H:%M:%S')
window_min = datetime.strptime('2021-02-10 12:00:00', '%Y-%m-%d %H:%M:%S')
```

```python
fig = plt.figure(figsize=(14,9))
plt.hist(df.created_at, bins=21, range=[window_max, window_min])
plt.ylabel('Rules created per day (linscale)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('over-replicated-rules.pdf')
plt.savefig('over-replicated-rules.png')

plt.grid(True, which="both", ls="-")
plt.ylabel('Rules created per day (logscale)')
plt.yscale('log')
plt.savefig('over-replicated-rules-log.pdf')
plt.savefig('over-replicated-rules-log.png')
```

```python
index = (df.created_at > datetime.strptime('2021-02-01 23:00:00', '%Y-%m-%d %H:%M:%S')) \
                 & (df.rse_expression != 'T1_US_FNAL_Disk')
```

```python
rul = df.rule_id[index]

for r in rul:
    print(r)
```

```python
nam = df.name[index]
nam = list(nam)
```

```python
for n in nam:
    print(n)
    counter = df.rse_expression[df.name == n].count()
    print("Rules for this dataset: {}".format(counter))
    
    rule_list = list(df.rule_id[df.name == n])
    express_list = list(df.rse_expression[df.name == n])
    date_list = list(df.created_at[df.name == n])

    for e,d,r in zip(express_list, date_list, rule_list):
        print("{} {}\t {}".format(d, r, e) )

        
    
    print()
    
    
```
